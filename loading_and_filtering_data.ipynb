```python
import pandas as pd
```

```python
filename = '/content/RailsToTrails_National_Count_Data_week.xlsx'
bike_df = pd.read_excel(filename, sheet_name = 'Bike Counts (14 counters)')
bike_df
```

```python
date_filtered = bike_df['Week of'] >= '2020-8-1'
bike_df[date_filtered]
```

```python
change_filtered = bike_df['Change 2019-2020'] > 1
bike_df[change_filtered]
```

```python
target_data = bike_df.loc[date_filtered & change_filtered, :]
target_data
```
